```python
#removal of correlated columns-feature reduction--feature >=0.8 correlation are removed
columns = np.full((corr.shape[0],), True, dtype=bool)
for i in range(x.shape[0]):
    for j in range(i+1, corr.shape[0]):
        if corr.iloc[i,j] >= 0.8:
            if columns[j]:
                columns[j] = False
```

```python
#filtered columns are the assigned to variable selected_columns

selected_columns = x.columns[columns]
print("the number of  filtered columns from correlation matrix are",selected_columns.shape)
```

```python
#random forest model
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score,confusion_matrix
from sklearn.metrics import accuracy_score

# It is found that we 92% data in train set and 8% data in test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.08, random_state=42)

#random forest classifier with n_estimators=10 (default)
clf_rf = RandomForestClassifier(random_state=20)      
clr_rf = clf_rf.fit(x_train,y_train)


#getting the accuracy score
ac = accuracy_score(y_test,clf_rf.predict(x_test))
print('Accuracy is: ',ac)
cm = confusion_matrix(y_test,clf_rf.predict(x_test))
sns.heatmap(cm,annot=True,fmt="d")
plt.savefig('RF_model.png')
```

```python
#selecting the top 20 features 
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

```

```python
# find best scored 20 features
select_feature = SelectKBest(chi2, k=20).fit(x_train, y_train)
print('Score list:', select_feature.scores_)
print('Feature list:', x_train.columns)
```

```python
x_train_2 = select_feature.transform(x_train)
x_test_2 = select_feature.transform(x_test)
#random forest classifier with n_estimators=10 (default)
clf_rf_2 = RandomForestClassifier()      
clr_rf_2 = clf_rf_2.fit(x_train_2,y_train)
ac_2 = accuracy_score(y_test,clf_rf_2.predict(x_test_2))
print('Accuracy is: ',ac_2)
cm_2 = confusion_matrix(y_test,clf_rf_2.predict(x_test_2))
sns.heatmap(cm_2,annot=True,fmt="d")
plt.savefig('RF_2_model.png')
```

```python
#for fetching the value of top 20 col names
from sklearn.feature_selection import RFE
# Create the RFE object and rank each pixel
clf_rf_3 = RandomForestClassifier()      
rfe = RFE(estimator=clf_rf_3, n_features_to_select=20, step=1)
rfe = rfe.fit(x_train, y_train)
print('Chosen best 20 feature by rfe:',x_train.columns[rfe.support_])
```

```python
# x and y are assigned
y = data_tt.xAttack                    # M or B 
x = data_tt.iloc[:,:-1]
```

```python
from keras.utils import np_utils
encoder=LabelEncoder()
encoder.fit(y)
y=encoder.transform(y)
y
```

```python
import numpy as np
from keras.utils import to_categorical

def encode(y):
    print('Shape of data (BEFORE encode): %s' % str(y.shape))
    encoded = to_categorical(y)
    print('Shape of data (AFTER  encode): %s\n' % str(encoded.shape))
    return encoded
```

```python
dummy_y = encode(y)
print(dummy_y)
```

```python
# split data
X_train, X_test, y_train, y_test = train_test_split(X_, dummy_y, test_size = 0.08, random_state = 0)
```

```python
X_train.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
classifier = Sequential() # Initialising the ANN

classifier.add(Dense(units = 100, activation = 'relu', input_dim =20 ))
classifier.add(Dense(units = 5,activation = 'softmax'))


# compile ANN
classifier.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])

# Fitting the data
hisroy =classifier.fit(X_train, y_train, batch_size = 55, epochs = 30)
```

```python
predictions = classifier.predict(X_test)
print(predictions.shape)
print(y_test.shape)
print(predictions)
```

```python
def decode(predictions):
    return np.argmax(predictions)
```

```python
for i in range(predictions.shape[0]):
    datum = predictions[i]
    print('index: %d' % i)
    print('encoded datum: %s' % datum)
    decoded_datum = decode(predictions[i])
    print('decoded datum: %s' % decoded_datum)
    print()
```
